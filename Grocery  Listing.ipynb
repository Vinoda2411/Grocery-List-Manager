# Grocery  Listing

```python
def display_menu():
    print("\n--- Grocery List Menu ---")
    print("1. View Grocery List")
    print("2. Add Item")
    print("3. Remove Item")
    print("4. Update Item Quantity")
    print("5. Clear List")
    print("6. Exit")

def view_list(grocery_list):
    if not grocery_list:
        print("Your grocery list is empty.")
    else:
        print("\nYour Grocery List:")
        for item, qty in grocery_list.items():
            print(f"- {item}: {qty}")

def add_item(grocery_list):
    item = input("Enter the item name to add: ").strip()
    qty = input("Enter quantity (default 1): ").strip()
    qty = int(qty) if qty.isdigit() else 1
    if item in grocery_list:
        grocery_list[item] += qty
    else:
        grocery_list[item] = qty
    print(f"Added {qty} x {item} to your list.")

def remove_item(grocery_list):
    item = input("Enter the item name to remove: ").strip()
    if item in grocery_list:
        del grocery_list[item]
        print(f"Removed {item} from your list.")
    else:
        print(f"{item} not found in the list.")

def update_item(grocery_list):
    item = input("Enter the item name to update: ").strip()
    if item in grocery_list:
        qty = input("Enter the new quantity: ").strip()
        if qty.isdigit():
            grocery_list[item] = int(qty)
            print(f"Updated {item} quantity to {qty}.")
        else:
            print("Invalid quantity entered.")
    else:
        print(f"{item} not found in the list.")

def clear_list(grocery_list):
    grocery_list.clear()
    print("Grocery list cleared.")

def main():
    grocery_list = {}
    while True:
        display_menu()
        choice = input("Choose an option (1-6): ").strip()
        if choice == '1':
            view_list(grocery_list)
        elif choice == '2':
            add_item(grocery_list)
        elif choice == '3':
            remove_item(grocery_list)
        elif choice == '4':
            update_item(grocery_list)
        elif choice == '5':
            clear_list(grocery_list)
        elif choice == '6':
            print("Exiting... Goodbye!")
            break
        else:
            print("Invalid choice. Please choose between 1-6.")

if __name__ == "__main__":
    main()
    
    
```
